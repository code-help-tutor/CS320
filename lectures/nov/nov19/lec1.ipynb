## Where do numpy arrays show up in ML?

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame([[0,2,1], [2,3,4], [8,5,6]], columns=["x1", "x2", "y"])
df
```

```python
df.values
```

```python
type(df.values)
```

```python
model = LinearRegression()
model.fit(df[["x1", "x2"]], df["y"])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(df[["x1", "x2"]])
```

```python
df[["x1", "x2"]].values @ model.coef_ + model.intercept_ # "@" is dot product -- to be learned soon!
```

## Vector Multiplication: Overview

### Elementwise Multiplication

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
*
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

### Dot Product

$\begin{bmatrix}
1 & 2 & 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 \\ 5 \\ 6
\end{bmatrix}$

$\begin{bmatrix}
1 \\ 2 \\ 3
\end{bmatrix}
\cdot
\begin{bmatrix}
4 & 5 & 6
\end{bmatrix}$

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
v1 * v2    # [1*4, 2*5, 3*6]
```

## Transpose

```python
v1 = np.array([1,2,3]).reshape(-1,1)
v2 = np.array([4,5,6]).reshape(-1,1)
v1 * v2
```

```python
v1.T * v2.T
```

## Broadcast

```python
v1.T.shape
```

```python
v2.shape
```

```python
v1.T * v2
```

```python
v1_T_broadcast = np.concatenate([v1.T, v1.T, v1.T])
v1_T_broadcast
```

```python
v2_broadcast = np.concatenate([v2, v2, v2], axis=1) # 0 means down (default), 1 means across
v2_broadcast
```

```python
v1_T_broadcast * v2_broadcast
```

```python
digits = np.arange(1, 10).reshape(-1,1)
digits
```

```python
digits.T * digits
```
