```python
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
```

```python
# example: a counters that ticks up

fig, ax = plt.subplots()

txt = ax.text(0.5, 0.5, "TODO", 
              size=48, va="center", ha="center")

def draw_frame(curr_frame_num):
    # way 1:
    #ax.cla() # clears the image before drawing the frame
    
    # way 2:
    txt.set_text(str(curr_frame_num))
    

# 1s (1000ms) between frames
anim = FuncAnimation(fig, draw_frame, frames=10, interval=1000)
html = anim.to_html5_video()
plt.close() # closes the figure

print(type(html))
HTML(html)
```

```python
with open("video.html", "w") as f:
    f.write(html)
```

```python
# example: a ball that rolls across the screen

fig, ax = plt.subplots()
ax.set_aspect("equal")

ball = plt.Circle((0, 0.5), 0.1)
ax.add_artist(ball)

platform = plt.Rectangle((0, 0.2), 1, 0.2, color="black")
ax.add_artist(platform)

# import frame-related variables
total_seconds = 3
fps = 30 # how many frames per second
frames = total_seconds * fps
interval = (1/fps) * 1000 # in milliseconds

def draw_frame(curr_frame_num):
    percent = curr_frame_num / frames # how much of the video is finished
    
    ball.center = (percent, 0.5)
    

anim = FuncAnimation(fig, draw_frame, frames=frames, interval=interval)
html = anim.to_html5_video()
plt.close() # closes the figure

print(type(html))
HTML(html)
```
