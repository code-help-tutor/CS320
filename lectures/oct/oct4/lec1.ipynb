## Paths

A **path** is a sequence of edges to get from one node to another

````python
# path 1: B => A => E
# path 2: B => C => D => E
# upcoming lectures: what is the shortest path?
# is a graph connected?  yes if path between any pair of nodes
g = Graph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "E")
g.edge("E", "A")
g.edge("X", "Y")
g
````

## Directed Graphs, DAGs

````python
from graphviz import Digraph
````

````python
g = Digraph()
g.edge("B", "A")
g
````

````python
g = Digraph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "E")
g.edge("E", "C")
g
# not connected: can't get from D to A, for example
# it is weakly connected: it would be connected if direction of edges were ignored
# A is a parent, B is a child (because A points to B)
````

````python
# DAG:
# directed
# acyclic
# graph
````

````python
g = Digraph() # not a DAG, because it has a cycle (not "acyclic")
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "A")
g
````

````python
g = Digraph() # it IS a DAG (no cycles)
g.edge("A", "B")
g.edge("C", "B")
g.edge("C", "A")
g
````

## Trees

**Tree**: DAG where one node (the **root**) has no parents and all others have exactly one parent

````python
# root: no parents
# leaves: no children
g = Digraph() # it IS a DAG (no cycles)
g.edge("C", "B")
g.edge("C", "A")
g.edge("A", "D")
g
````

````python
# linked list: a tree where each node has at most one child
# binary tree: a tree where each node has at most two children

# for a binary tree
class Node:
    def __init__(self, name):
        self.name = name
        # self.children = [...other nodes...]
        #self.child = None
        
        self.left = None
        self.right = None
        
    def dump(self):
        print(self.name)
        if self.left:
            self.left.dump()
        if self.right:
            self.right.dump()
        
A = Node("A")
B = Node("B")
C = Node("C")
D = Node("D")
A.left = B
A.right = C
C.left = D

A.dump()
````
