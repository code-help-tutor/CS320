## Paths

A **path** is a sequence of edges to get from one node to another

````python
g = Graph()
g.edge("A", "B")
g.edge("B", "C")
g.edge("C", "D")
g.edge("D", "F")
g.edge("F", "A")
g.node("E")
g
# B->A->F is a path
# B->C->D->F is path
# what is the shortest path between two nodes?
# is there path?  (not between A and E)
````

````python
# connected: there is a path between any pair of nodes
````

## Directed Graphs, DAGs

````python
from graphviz import Digraph
````

````python
# B is a parent of A, A is a child of B
# C is a grandchild of B (A and C are both descendants of B)
g = Digraph()
g.edge("B", "A")
g.edge("A", "C")
g
````

````python
# B is a parent of A, A is a child of B
# C is a grandchild of B (A and C are both descendants of B)
g = Digraph()
g.edge("B", "A")
g.edge("A", "C")
g.edge("C", "B")
g # has a cycle!
````

````python
# a node can be it's own descendant if there are cycles (path from a node back to itself)!
````

````python
# B is a parent of A, A is a child of B
# C is a grandchild of B (A and C are both descendants of B)
g = Digraph()
g.edge("B", "A")
g.edge("B", "D")
g.edge("A", "C")
g.edge("D", "C")
g # not a cycle!
# is this a connected graph? no, but it is a DAG
# is this a "weakly" connected graph?  yes,
# because it would be connected if we ignored edge directions
````

## Trees

**Tree**: DAG where one node (the **root**) has no parents and all others have exactly one parent

````python
g = Digraph()
g.edge("B", "A")
g.edge("B", "D")
g.edge("D", "C")
g.edge("C", "E")
g
# root: no parents, leaf: no children
````

````python
# max one child per node (Linked List)
# max two children per node (Binary Tree)
````

````python
class Node:
    def __init__(self, name):
        self.name = name
        
        # in general
        #self.children = [....]
        
        # for a linked list
        # self.child = None # reference to another Node (an edge can be an attribute)
        
        # for a binary tree
        self.left = None
        self.right = None
        
A = Node("A")
B = Node("B")
C = Node("C")
A.left = B
A.right = C
````

````python
class Node:
    def __init__(self, name):
        self.name = name
        self.left = None
        self.right = None
        
    def dump(self): # print out all node names
        print(self.name)
        if self.left:
            self.left.dump()
        if self.right:
            self.right.dump()
        
A = Node("A")
B = Node("B")
C = Node("C")
D = Node("D")
A.left = B
A.right = C
C.right = D
A.dump()
````
