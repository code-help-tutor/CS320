```python
from graphviz import Digraph, Graph
from IPython.core.display import display
```

```python
# BST: Binary Search Tree

class Node:
    def __init__(self, name):
        self.name = name # use as key (if dict)
        # self.value = ???? (if dict)
        self.left = None
        self.right = None
        
    def draw_edges(self, g):
        g.node(str(self.name))
        if self.left != None:
            g.edge(str(self.name), str(self.left.name), label="L")
            self.left.draw_edges(g)
        if self.right != None:
            g.edge(str(self.name), str(self.right.name), label="R")
            self.right.draw_edges(g)
    
    def draw_graph(self):
        g = Digraph() # graphviz graph
        self.draw_edges(g)
        return g
    
    def _repr_svg_(self): # SVG is a common image format
        g = self.draw_graph()
        return g._repr_svg_()
    
    # for dict:
    # __getitem__
    # __setitem__
    
    def __contains__(self, x):
        #print(f"search subtree for {x}:")
        #display(self)
        if self.name == x:
            return True
        if x < self.name and self.left and self.left.__contains__(x):
            return True
        if x > self.name and self.right and self.right.__contains__(x):
            return True
        return False
    
    def add(self, x):
        if x < self.name:
            # GO LEFT
            if self.left == None:
                self.left = Node(x)
            else:
                self.left.add(x)
        elif x > self.name:
            # GO RIGHT
            if self.right == None:
                self.right = Node(x)
            else:
                self.right.add(x)
    
# BST rule: each node's value is BIGGER than all those in the LEFT subtree, and
#           each node's value is SMALLER than all those in the RIGHT subtree
values = [7, 3, 11, 15, 12, 20]
root = Node(values[0])
for v in values[1:]:
    root.add(v)
# root.__contains__(20)
19 in root
```
