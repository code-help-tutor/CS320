```python
# binary trees
from graphviz import Digraph
from IPython.core.display import display

# search:
# 3 in SOME_SET    ****
# SOME_DICT["key"]

class Node:
    def __init__(self, name):
        self.name = name
        self.left = None
        self.right = None
        
    def draw_edges(self, g):
        g.node(str(self.name))
        if self.left != None:
            g.edge(str(self.name), str(self.left.name), label="L")
            self.left.draw_edges(g)
        if self.right != None:
            g.edge(str(self.name), str(self.right.name), label="R")
            self.right.draw_edges(g)
    
    def draw_graph(self):
        g = Digraph() # graphviz graph
        self.draw_edges(g)
        return g
    
    def _repr_svg_(self): # SVG is a common image format
        g = self.draw_graph()
        return g._repr_svg_()
    
    def contains(self, x):
        print(f"search subtree for {x}:")
        display(self)
        if self.name == x:
            return True
        if self.left and self.left.contains(x):
            return True
        if self.right and self.right.contains(x):
            return True
        return False
        
root = Node(7)
root.left = Node(3)
root.left.right = Node(20)
root.right = Node(11)
root.right.left = Node(12)
root.right.right= Node(15)
#7 in root
root.contains(15) # returns T/F
# O(N) where N is the number of nodes
```

```python
# BST: Binary Search Tree

# BST rule: each node's value is BIGGER than all those in the LEFT subtree, and
#           each node's value is SMALLER than all those in the RIGHT subtree

class Node:
    def __init__(self, name):
        self.name = name
        self.left = None
        self.right = None
        
    def draw_edges(self, g):
        g.node(str(self.name))
        if self.left != None:
            g.edge(str(self.name), str(self.left.name), label="L")
            self.left.draw_edges(g)
        if self.right != None:
            g.edge(str(self.name), str(self.right.name), label="R")
            self.right.draw_edges(g)
    
    def draw_graph(self):
        g = Digraph() # graphviz graph
        self.draw_edges(g)
        return g
    
    def _repr_svg_(self): # SVG is a common image format
        g = self.draw_graph()
        return g._repr_svg_()
    
    def contains(self, x):
        print(f"search subtree for {x}:")
        display(self)
        if self.name == x:
            return True
        if x < self.name and self.left and self.left.contains(x):
            return True
        if x > self.name and self.right and self.right.contains(x):
            return True
        return False

root = Node(7)
root.left = Node(3)
root.right = Node(11)
root.right.right = Node(15)
root.right.right.left = Node(12)
root.right.right.right = Node(20)
root.contains(4)
```
