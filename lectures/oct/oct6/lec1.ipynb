```python
from graphviz import Digraph, Graph
from IPython.core.display import display

# binary trees: how can we use them for search?
#3 in SOME_SET
#SOME_DICT["key"]

class Node:
    def __init__(self, name):
        self.name = name
        self.left = None
        self.right = None
        
    def draw_edges(self, g):
        g.node(str(self.name))
        if self.left:
            g.edge(str(self.name), str(self.left.name), label="L")
            self.left.draw_edges(g)
        if self.right:
            g.edge(str(self.name), str(self.right.name), label="R")
            self.right.draw_edges(g)

    def draw_graph(self):
        g = Graph()
        self.draw_edges(g)
        return g
    
    def _repr_svg_(self): # svg is an image format
        g = self.draw_graph()
        return g._repr_svg_()
    
    def contains(self, x):
        print(f"Searching for {x} in subtree:")
        display(self)
        if self.name == x:
            return True
        if self.left and self.left.contains(x):
            return True
        if self.right and self.right.contains(x):
            return True
        return False
        
root = Node(5)
root.left = Node(3)
root.left.left = Node(20)
root.right = Node(7)
root.right.left = Node(8)
root.right.right = Node(11)
#3 in root
root.contains(11) # return T/F
```

```python
# Binary Search Trees: BST (for fast lookup)

# BST rule:
# each node's value is bigger than all those in it's left subtree, and
# each node's value is small than all those in it's right subtree

root = Node(5)
root.left = Node(3)
root.right = Node(7)
root.right.right = Node(11)
root.right.right.right = Node(20)
root.right.right.left = Node(8)


root
```
